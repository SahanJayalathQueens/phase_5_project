```python
# Extracting the last 5 digits from the address to create the zip_code column
df['zip_code'] = df['address'].str.extract(r'(\d{5})$')
# Rename the 'address/zipcode' column to match df
df2 = df2.rename(columns={'address/zipcode': 'zip_code'})
```

```python
df2.rename(columns={
    'address/city': 'city',                    # 'address/city' -> 'city'
    'address/streetAddress': 'address',        # 'address/streetAddress' -> 'address'
    'price': 'price',                          # 'price' -> 'price' (same in both)
    'bedrooms': 'bedroom_number',              # 'bedrooms' -> 'bedroom_number'
    'bathrooms': 'bathroom_number',            # 'bathrooms' -> 'bathroom_number'
    'livingArea': 'living_space',              # 'livingArea' -> 'living_space'
    'zip_code': 'zip_code'                     # 'zip_code' -> 'zip_code' (same in both)
}, inplace=True)
```

```python
# Drop unnecessary columns from df and df2
df.drop(columns=['street_name'], inplace=True)  # Drop 'street_name' from df
df2.drop(columns=['Unnamed: 0', 'yearBuilt'], inplace=True)  # Drop 'Unnamed: 0' and 'yearBuilt' from df2

# Check the results by printing the first few rows of both dataframes
print("df after dropping street_name:")
print(df.head())

print("\ndf2 after dropping Unnamed: 0 and yearBuilt:")
print(df2.head())
```

```python
# Convert the 'zip_code' column in df2 to integers
df2['zip_code'] = df2['zip_code'].astype('Int64')  # Using 'Int64' to handle any NaN values


```

```python
# Convert 'zip_code' in df to the same type as in df2 (Int64)
df['zip_code'] = df['zip_code'].astype('Int64')

# Now merge the two dataframes using 'outer' join
merged_df = pd.merge(df, df2, on=['city', 'address', 'price', 'bedroom_number', 'bathroom_number', 'living_space', 'zip_code'], how='outer')

# Check the resulting dataframe
print(merged_df.head())

# Optionally, check the shape of the resulting dataframe
print("Merged DataFrame shape:", merged_df.shape)
```

```python
merged_df
```

```python
# Check for missing values (NaNs) in the merged DataFrame
nan_counts = merged_df.isna().sum()

# Display the count of missing values for each column
print("Missing values per column:")
print(nan_counts)
```

```python
# Drop rows where 'price' or 'zip_code' are missing
merged_df_clean = merged_df.dropna(subset=['price', 'zip_code'])

# Check the result
print("Number of rows after dropping missing values:", merged_df_clean.shape[0])

# Display the first few rows to verify
print(merged_df_clean.head())
```

```python
# Export the cleaned DataFrame to a CSV file
merged_df_clean.to_csv('NYCH.csv', index=False)

print("File 'NYCH.csv' has been successfully saved.")
```
